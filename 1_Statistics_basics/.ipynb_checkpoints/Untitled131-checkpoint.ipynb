```python
#Задание 1. Загрузка данных

'''Изучить представленный набор данных на основе описания его столбцов, 
загрузить его и оставить 8 столбцов для дальнейшего изучения: 
surgery?, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

'''


import pandas as pd
import numpy as np

df = (pd.read_csv('C:\\Users\\Оляша\\Desktop\\Netology\\DZ\\3_Statistics\\1_Statistics_basics\\horse_data.csv', 
                  header = None, usecols=[0, 1, 3, 4, 5, 6, 10, 22]))

df.columns = ['surgery?', 'Age', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain', 'outcome'] 

     

df.head(10)
```

```python
#Задание 2. Первичное изучение данных

'''Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.'''

'''При анализе значений по столбцам обнаружены пропуски данных, обозначенные символами "?", 
а значит необходимо от них избавиться (замена на None)'''

def change_item(item):
    if item=='?':
        item = np.NaN
    return item


df =  df.applymap(change_item)

df.info()

'''
Обнаружено, что во всех столбцах, кроме Age, данные нечисловые, следовательно, 
для дальнейшего расчета базовых статистик необходимо это исправить, например, с помощью метода to_numeric 
(значения None будут считаться отсутствующими)
'''

df_numeric = df.apply(pd.to_numeric, errors='coerce') 

#Рассчет базовых статистик

median = df_numeric.median().to_frame().unstack().unstack()#переводим Series в df и транспонируем 
median.index = ['median']
mode = df_numeric.mode()
mode.index = ['mode']
variance = df_numeric.var().to_frame().unstack().unstack()
variance.index = ['variance']

#Свод базовых статистик в один датафрейм для наглядности
df_stat0 = pd.concat([df_numeric.describe(), median, mode,variance]) 

df_stat0
```

```python
#Поиск выбросов через межквартальный размах
#Расчет границ для определения выбросов

df_stat0.loc['lower_bound ',] = df_stat0.loc['25%', ] - 1.5*(df_stat0.loc['75%', ] - (df_stat0.loc['25%', ]))
df_stat0.loc['upper_bound ',] = df_stat0.loc['75%', ] + 1.5*(df_stat0.loc['75%', ] - (df_stat0.loc['25%', ]))

df_stat0
```

```python
#Задание 3. Работа с пропусками
"""Рассчитать количество пропусков для всех выбранных столбцов. 
Принять и обосновать решение о методе заполнения пропусков по каждому столбцу 
на основе рассчитанных статистик и возможной взаимосвязи значений в них. 
Сформировать датафрейм, в котором пропуски будут отсутствовать"""


#Обработаем пропуски в датафреме по жеребятам
df_foals.isna().sum() #количество NaN значений в датафрейме по жеребятам (Age = 9)
#(df_numeric.isna().mean() * 100).round(2) #процентное отношение NaN к общему числу элементов

#df_numeric[df_numeric.isnull().any(1)]

```

```python
'''
Исключим строки, которые содержат пропуски по столбцам: 
rectal temperature,pulse, respiratory rate, temperature of extremities, 
поскольку одновременное отсутствие этих данных приводит к потере информативности всего представленного набора данных
и невозможности заполнить пропуски на основе корреляции. 
Кроме того, указанные строки не окажут влияния на расчет базовых статистик указанных столбцов, 
содержащих наиболее значимые характеристики для рассматриваемого набора данных
'''
df_clean_foals = df_foals.copy().dropna(how = 'all', subset=['rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities'])

#df_numeric[df_numeric['rectal temperature'].isnull()& df_numeric['pulse'].isnull()&df_numeric['respiratory rate'].isnull()&df_numeric['temperature of extremities'].isnull()]
#df_clean_foals[df_clean_foals.isnull().any(1)]


#Поскольку значения respiratory rate имеют высокие колебания, то заменим пропуски в данном столбце на медиану, 
#чтобы минимизировать влияние на базовые статистики

df_clean_foals['respiratory rate'].fillna(df_clean_foals['respiratory rate'].median(), inplace=True)



#заменим пропуски в столбце rectal temperature на медиану, чтобы минимизировать влияние на базовые статистики

df_clean_foals['rectal temperature'].fillna(df_clean_foals['rectal temperature'].median(), inplace=True)

#df_clean_foals.groupby(['rectal temperature', 'temperature of extremities']).count()

#Поскольку rectal temperature и temperature of extremities коррелированы, 
#то пропуски temperature of extremities можно заменить исходя из имеющихся значений rectal temperature


#df_clean_foals["temperature of extremities"].fillna(df_clean_foals.groupby(["rectal temperature"])["temperature of extremities"].transform(lambda x: x.mode()[0]))

df_clean_foals



#df_clean_foals_groups["temperature of extremities"].fillna(method='bfill', inplace=True)
#df_clean_foals_groups
```

```python
#Обработаем пропуски в датафреме по лошадям


df_horse.isna().sum() #количество NaN значений в датафрейме по лошадям (Age = 1)
df_horse.info()

df_horse
```

```python
#Воспользуемся тем фактом, что столбцы rectal temperature и temperature of extremities коррелированы (исходя из описания данных)
#Для начала заполним NaN в столбце temperature of extremities.
# Предположим, что средние значения temperature of extremities различаются в зависимости от того, 
# была ли операция или нет


print(df_horse.groupby('surgery?')['temperature of extremities'].median())
df_temp_extr = df_horse.copy()
df_temp_extr['temperature of extremities'].fillna(df_horse.groupby('surgery?')['temperature of extremities'].transform('median'), inplace=True)
```

```python
#Воспользуемся тем фактом, что столбцы rectal temperature и temperature of extremities коррелированы (исходя из описания данных)
#Определим моду среди значений rectal temperature исходя из temperature of extremities

df_rect_temp_gr = df_temp_extr.groupby([ 'temperature of extremities'])['rectal temperature'].agg(pd.Series.mode).to_frame().reset_index()
df_rect_temp_gr.columns=['temperature of extremities', 'rectal temperature']
df_rect_temp_gr['rectal temperature'] = df_rect_temp_gr['rectal temperature'].map(lambda x: x[0] if x.size > 1 else x)
df_rect_temp_gr


#Более красивое решение
# from scipy import stats
# import scipy

# df_horse1 = df_horse.groupby([ 'temperature of extremities'])['rectal temperature'].agg(lambda x: scipy.stats.mode(x)[0])
# df_horse1.reset_index() 
    
```

```python
#Отфильтруем строки, в которых значения rectal temperature равны NaN,а temperature of extremities не NaN
df_horse_rect_temp_nun = (df_temp_extr.loc[df_temp_extr['rectal temperature'].isnull() 
                                       & df_temp_extr['temperature of extremities'].notnull()])

#Сольем таблицы (сохраняя индекс), чтобы заменить NaN по rectal temperature на моду исходя из temperature of extremities
df_horse_rect_temp_merge = (df_horse_rect_temp_nun.reset_index().
                      merge(df_rect_temp_gr, how="left", on = 'temperature of extremities').set_index('index'))

         
#Заменим NaN по rectal temperature в общем датафрейме по лошадям
df_horse_rect_temp = df_temp_extr.copy()
df_horse_rect_temp['rectal temperature'].fillna(df_horse_rect_temp_merge['rectal temperature_y'], inplace = True)
df_horse_rect_temp
```

```python
#Исходя из описания столбцов 'surgery?' и 'pain',
#а именно принимая тот факт, что чем больнее, тем больше вероятность того, что потребуется операция, 
#определим моду среди значений 'surgery?' исходя из 'pain' 
#и исходя из полученных данных заполним NaN в указанных столбцах

from scipy import stats
import scipy

df_horse_surg_gr = df_horse_rect_temp.groupby([ 'pain'])['surgery?'].agg(lambda x: scipy.stats.mode(x)[0])
df_horse_surg_gr.reset_index() 
    
```

```python
#Отфильтруем строки, в которых значения surgery?  равны NaN, а pain нет
df_horse_surg_nun = (df_horse_rect_temp.loc[df_horse_rect_temp['surgery?'].isnull() 
                                       & df_horse_rect_temp['pain'].notnull()])

#Сольем таблицы (сохраняя индекс), чтобы заменить NaN по surgery? на моду surgery? исходя из pain
df_horse_surg_merge = (df_horse_surg_nun.reset_index().
                      merge(df_horse_surg_gr, how = "left", on = 'pain').set_index('index'))

        
#Заменим NaN по surgery? в общем датафрейме по лошадям
df_horse_surg = df_horse_rect_temp.copy()
df_horse_surg['surgery?'].fillna(df_horse_surg_merge['surgery?_y'], inplace = True)
df_horse_surg
```

```python
#определим моду среди значений 'pain' исходя из 'surgery?' 
#и исходя из полученных данных заполним NaN в указанных столбцах

df_horse_pain_gr = df_horse_surg.groupby(['surgery?'])['pain'].agg(lambda x: scipy.stats.mode(x)[0])
df_horse_pain_gr.reset_index() 
```

```python
#Отфильтруем строки, в которых значения pain  равны NaN, а surgery? нет
df_horse_pain_nun = (df_horse_surg.loc[df_horse_surg['pain'].isnull() 
                                       & df_horse_surg['surgery?'].notnull()])

#Сольем таблицы (сохраняя индекс), чтобы заменить NaN по pain на моду pain исходя из surgery?
df_horse_pain_merge = (df_horse_pain_nun.reset_index().
                      merge(df_horse_pain_gr, how = "left", on = 'surgery?').set_index('index'))

       
#Заменим NaN по pain в общем датафрейме по лошадям
df_horse_pain = df_horse_surg.copy()
df_horse_pain['pain'].fillna(df_horse_pain_merge['pain_y'], inplace = True)
df_horse_pain
```

```python
#Попробуем найти зависимость между столбцами 'surgery?' и 'outcome'.
#Cоставим кросс-таблицу по соотношению значений 'surgery?' и 'outcome' 
#и исходя из полученных данных заполним NaN в указанных столбцах

df_horse_gr_outcome = pd.crosstab(df_horse['surgery?'], df_horse['outcome'])
df_horse_gr_outcome

#поскольку частота значений 1.0 в столбце outcome высока при любых значениях столбца surgery?, 
#то заполним  NaN в outcome  значениями 1.0
```

```python
#Для начала заполним NaN в столбце temperature of extremities.
df_foals_temp_extr = df_foals.copy()
df_foals_temp_extr['temperature of extremities'].fillna(df_foals.groupby('surgery?')['temperature of extremities'].transform('median'), inplace=True)

#Определим моду среди значений rectal temperature исходя из temperature of extremities
df_foals_rect_temp_gr = df_foals.groupby([ 'temperature of extremities'])['rectal temperature'].agg(lambda x: scipy.stats.mode(x)[0])
df_foals_rect_temp_gr.reset_index() 

#Отфильтруем строки, в которых значения rectal temperature равны NaN,а temperature of extremities не NaN
df_foals_rect_temp_nun = (df_foals_temp_extr.loc[df_foals_temp_extr['rectal temperature'].isnull() 
                                       & df_foals_temp_extr['temperature of extremities'].notnull()])

#Сольем таблицы (сохраняя индекс), чтобы заменить NaN по rectal temperature на моду исходя из temperature of extremities
df_foals_rect_temp_merge = (df_foals_rect_temp_nun.reset_index().
                      merge(df_foals_rect_temp_gr, how="left", on = 'temperature of extremities').set_index('index'))

        
#Заменим NaN по rectal temperature в общем датафрейме по лошадям
df_foals_rect_temp = df_foals_temp_extr.copy()
df_foals_rect_temp['rectal temperature'].fillna(df_foals_rect_temp_merge['rectal temperature_y'], inplace = True)
df_foals_rect_temp

#определим моду среди значений 'pain' исходя из 'surgery?' 
#и исходя из полученных данных заполним NaN в указанных столбцах

df_foals_pain_gr = df_foals_rect_temp.groupby(['surgery?'])['pain'].agg(lambda x: scipy.stats.mode(x)[0])
df_foals_pain_gr.reset_index() 

#Отфильтруем строки, в которых значения pain  равны NaN, а surgery? нет
df_foals_pain_nun = (df_foals_rect_temp.loc[df_foals_rect_temp['pain'].isnull() 
                                       & df_foals_rect_temp['surgery?'].notnull()])

#Сольем таблицы (сохраняя индекс), чтобы заменить NaN по pain на моду pain исходя из surgery?
df_foals_pain_merge = (df_foals_pain_nun.reset_index().
                      merge(df_foals_pain_gr, how = "left", on = 'surgery?').set_index('index'))

       
#Заменим NaN по pain в общем датафрейме по жеребятам
df_foals_pain = df_foals_rect_temp.copy()
df_foals_pain['pain'].fillna(df_foals_pain_merge['pain_y'], inplace = True)
df_foals_pain

#Оставшиеся NaN заменим на медианы соответствующих столбцов, чтобы минимизировать влияние на базовые статистики

# Предположим, что средние значения пульса различаются в зависимости от боли 
#https://www.equinevet.ru/post/%D1%84%D0%B8%D0%B7%D0%B8%D0%BE%D0%BB%D0%BE%D0%B3%D0%B8%D1%87%D0%B5%D1%81%D0%BA%D0%B8%D0%B5-%D0%BF%D0%BE%D0%BA%D0%B0%D0%B7%D0%B0%D1%82%D0%B5%D0%BB%D0%B8-%D0%BB%D0%BE%D1%88%D0%B0%D0%B4%D0%B8

df_foals_pulse = df_foals_pain.copy()
df_foals_pulse['pulse'].fillna(df_foals_pain.groupby('pain')['pulse'].transform('median'), inplace=True)
df_foals_pulse

#Предположим, что средние значения частоты дыхания также различаются в зависимости от боли 

df_foals_resp_rate = df_foals_pulse.copy()
df_foals_resp_rate['respiratory rate'].fillna(df_foals_pulse.groupby('pain')['respiratory rate'].transform('median'), inplace=True)
df_foals_resp_rate

#Поскольку моды среди rectal temperature для значения 2 у temperature of extremities в данной выборке нет, 
#то заполним оставшееся значение NaN в rectal temperature  медианой по данному столбцу
df_foals_resp_rate1 = df_foals_resp_rate.copy()
df_foals_resp_rate1['rectal temperature'].fillna(df_foals_resp_rate['rectal temperature'].median(), inplace=True)
df_foals_resp_rate1 
```

```python
#Таким образом значения NaN в исходном датафрейме заполнены

df_end = pd.concat([df_horse_resp_rate, df_foals_resp_rate1]) 
df_end.info() 
df_end.describe() 

median = df_end.median().to_frame().unstack().unstack()#переводим Series в df и транспонируем 
median.index = ['median']
mode = df_end.mode()
mode.index = ['mode']
variance = df_end.var().to_frame().unstack().unstack()
variance.index = ['variance']

#Свод базовых статистик в один датафрейм для наглядности
df_stat_end = pd.concat([df_end.describe(), median, mode,variance]) 

df_stat_end
```
